```python
import numpy as np
a= np.array([1,2,3])
print(a)
```

```python
arr=np.array([1,2,3,4,5])
print(arr)
```

```python
arr=np.array([[1,2,3],[4,5,6]])
print(arr.ndim) #dimension
print(arr.shape) #rows and columns
print(arr.size) #ele num
print(arr.dtype) # data types
```

```python
#indexing and slicing
#1D indexing
arr=np.array([10,20,30,40])
print(arr[2])
#2D indexing
arr=np.array([[1,2,3],[4,5,6]])
print(arr[1,2])
```

```python
#slicing
b=np.array([10,20,30,40,50])
print(b[2:4])
```

```python
#operation
ar=np.array([10,20,30,40,50])
print(ar[1] * 10 *ar[3])
print(ar[3] +3% ar[4])
```

```python
#for 1D array
a1=np.array([1,2])
a2=np.array([3,4])
a=np.concatenate([a1,a2])
print(a)
#for 2D array
b=np.array([[1,2]])
c=np.array([[3,4]])
print(np.concatenate((b,c),axis=0)) #rowwise
print(np.concatenate((b,c),axis=1))
```

```python
#reshaping array
import numpy as np

arr = np.array([1,2,3,4,5,6])
print(arr.reshape(3,2))
```

```python
#splitting arrays
arr=np.array([10,20,30,40,50,60])
print(np.split(arr,3))
```
